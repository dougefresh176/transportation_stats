```python
residence_vs_vehicles = pd.pivot_table(cars_years,'Workers in Households','Vehicles Available 6','Length of US Residence 6',aggfunc='sum')
```

```python
residence_vs_vehicles = residence_vs_vehicles.rename(columns={'Born in the United States, Puerto Rico or the Island Areas, or Born Abroad to U.S. Citizen Parent(s)':'Born & lives elsewhere'})
```

```python
 
residence_vs_vehicles[['In US 10 or more years',
 'In the US 2 years or less',
 'In the US 3 to 5 years',
 'In the US 6 to 9 years']].plot(kind='bar',figsize=(10,6))
```

```python
residence_vs_vehicles = residence_vs_vehicles.reindex(columns=['In the US 2 years or less','In the US 3 to 5 years','In the US 6 to 9 years','In US 10 or more years','Born in the United States, Puerto Rico or the Island Areas, or Born Abroad to U.S. Citizen Parent(s)','Total, length of US residence']) 
```

```python
vehicles_vs_residence = residence_vs_vehicles.swapaxes(axis1="index",axis2="columns")
```

```python
vehicles_vs_residence.iloc[:4].plot(kind='bar',figsize=(10,6))
```
